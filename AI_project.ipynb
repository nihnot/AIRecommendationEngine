```python
import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split

dfRatings = pd.read_csv("ratings.csv")
dfMovies = pd.read_csv("movies.csv")

reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(dfRatings[['userId', 'movieId', 'rating']], reader)

trainset, testset = train_test_split(data, test_size=0.2)

model = SVD()

model.fit(trainset)

predictions = model.test(testset)

def recommend_items_for_user(userId, model, dfMovies, num_recommendations=10):
    unrated_items = set(dfMovies['title']) - set(dfRatings[dfRatings['userId'] == userId]['movieId'])
    predictions = [(item, model.predict(userId, item).est) for item in unrated_items]
    recommendations = sorted(predictions, key=lambda x: x[1], reverse=True)[:num_recommendations]
    return [item[0] for item in recommendations]


full_trainset = data.build_full_trainset()
full_model = SVD()
full_model.fit(full_trainset)

user_id_to_recommend = "3"
recommended_items = recommend_items_for_user(user_id_to_recommend, full_model, dfMovies)
print(f"Recommended items for user '{user_id_to_recommend}':")
print(recommended_items)
```
